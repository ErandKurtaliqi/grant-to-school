```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, Binarizer
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('dataset/grant-to-school.csv')
df.head()
```

```python
#mbledhja e informacionit rreth datasetit
print(" Dataset-i u lexua me sukses!\n")
print("Numri i rreshtave dhe kolonave ne dataset eshte:", df.shape)
print("\nShembull nga të dhënat:\n", df.head())
    
```

```python
#definimi i tipevete te te dhenave 
print("\n Informacion mbi kolonat:")
print(df.info())

#konvertimi i tipeve te te dhenave
df["Year"] = df["Year"].astype(int)
df["Amount of Scholarship"] = pd.to_numeric(df["Amount of Scholarship"], errors="coerce")
```

```python
#kontrolli i vlerave bosh
print("\n Vlerat e zbrazëta për kolonë:")
print(df.isnull().sum())

# Zëvendëso vlerat bosh në kolona numerike me mesatare
df["Amount of Scholarship"].fillna(df["Amount of Scholarship"].mean(), inplace=True)

# Zëvendëso vlerat bosh në kolona kategorike me modë (vlera më e shpeshtë)
for col in ["Gender", "State name", "District name", "Category name", "Class"]:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python

#heqja e rreshtave te dyfishte dhe pastrimi i hapesirave te panevojshme
df = df.drop_duplicates()

for col in ["State name", "District name", "Category name", "Gender"]:
    df[col] = df[col].astype(str).str.strip()

print("\n Pastrimi përfundoi! Rreshta pas pastrimit:", len(df))
```
